## ZIP Function
- zip function returns a zip object which is an iterator of two or more iterables
- zip funcion enables us to loop on many iterables in the same loop

**which iterable controls the number of iterations performed:**
- the iterable with least number of elements

```python
list_1 = [1,2,3,4,5,6]

list_2 = ['a','b','c','d','e','f']

tuple_1 = (1,'r',66,'data',1.4)

dict_1 = {"name": 'ahmed', "country": "Egypt",'age':18, 'job': "video editor"}
```

```python
zip_1 = zip(list_1, list_2)

print(zip_1)
print(type(zip_1))
```

to verify that the zip object is an iterator use the next function on it which works only on iterator objects

```python
next(zip_1)
```

```python
print(type(next(zip_1)))
```

to print the items in the zip object we have to loop over it

```python
for item in zip_1:
    print(item)
```

```python
zip_2 = zip(list_2, dict_1)

for item in zip_2:   #dict will control as it has less number of items
    print(item) # print a tuple of characters from the tuples, and keys from the dict
```

```python
# to print the values in the dict 

for tuple_item, dict_item in zip_2 :
    print (tuple_item, "             ", dict_item)
```

```python
zip_3 = zip(list_2, tuple_1, tuple_1)

for i1, i2, i3 in zip_3 :
    print (i1, "  ", i2, "  ",i3, "   " ,dict_1[i3])
```

to loop over many iterables using foor loops

```python
%%timeit
n = min (len(list_1), len(list_2), len(tuple_1))

for i in range(n):
    print (list_1[i] , list_2[i], tuple_1[i])
```
